```r
library('readr')
library('magrittr')
library('dplyr')
library('tidyr')
library('ggplot2')
```

```r
tsubsets <- read_csv(
    "data.csv",
    col_types=cols(
        `Donor`=col_factor(levels=paste("Donor", c(10, 12, 14, 15))),
        `Day`=col_factor(levels=1:6),
        `Condition`=col_factor(levels=c("No RNA", "1600 ng", "800 ng", "400 ng")),
        `Statistic`=col_factor(levels=c("GFP+")),
        `Value`=col_double()
    )
)
```

```r
tsubsets %>%
    ggplot(aes(x=`Day`, y=`Value`, group=`Condition`, color=`Condition`)) + 
        geom_point() + 
        geom_line() +
        ylab('GFP Positive Percent (%)') +
        facet_wrap(~Donor, ncol=4)
```
